This code is to gather all the common names in one scientific names and to don't have duplicate scientific names.

```python
species_names_unique = species_info.scientific_name.unique()
species_common_names_unique = []
category_unique = []
conservation_status_unique = []
for specie in species_names_unique:
   x = species_info[ species_info.scientific_name == specie ]
   category_unique.append( np.unique(x.category.values)[0] )
   conservation_status_unique.append( np.unique(x.conservation_status.values)[0] ) 
   y = []
   for jj in x.common_names:
        z = jj.split(', ') 
        y.append( z )
   y = np.concatenate(y)
   y = np.unique(y)
   species_common_names_unique.append(y) 

species_info_unique = pd.DataFrame( { 'category' : category_unique, 'scientific_name' : species_names_unique,
                                  'common_names' : species_common_names_unique, 
                                     'conservation_status' : conservation_status_unique} )
```

Now, we have a table with unique scientific names.

```python
species_info_unique
```

To left join the observations and species_info_unique tables the pandas merge function is used.

```python
data = pd.merge(observations, species_info_unique, on = 'scientific_name', how = 'left',
                indicator = True, validate='many_to_one')
data
```

what proportion of each category are in each conservcation status?
Grouping by catgeory and conservation status and counting the spcecies gives the data_by_conservation_and_category table.

```python
data_by_conservation_and_category = data.groupby(['category','conservation_status']).scientific_name.count().reset_index()
data_by_conservation_and_category.rename(columns = {'scientific_name' : 'species_count'}, inplace = True)
data_by_conservation_and_category
```

```python
categories = data_by_conservation_and_category.category.unique().tolist()
categories
```

Calculating the percentage of species by category. The sum of the percentages of each category is printed to check.

```python
species_count = [  ]
for jj in categories:
    x = data_by_conservation_and_category[ data_by_conservation_and_category.category == jj ].species_count.sum()
    y = 100.0*data_by_conservation_and_category[ data_by_conservation_and_category.category == jj ].species_count/x 
    species_count.append(y)
    print( jj, y.sum() )

species_count = np.concatenate(species_count)
```

adding the percentage colum to data_by_conservation_and_category

```python
data_by_conservation_and_category['percentage'] = species_count
```

### Which are the most and the least observed species in each park?

We define a function to be used to show the most and the least observed species in each park.

```python
def get_min_max_observations(dataF, what):
   df = pd.DataFrame() 
   for park in dataF.park_name.unique():
      data_by_park = dataF[ dataF.park_name == park ].reset_index(drop=True)
      if ( what == 'min'):
         df = df.append(data_by_park[ data_by_park.observations == data_by_park.observations.min() ] ).reset_index(drop=True)
      elif ( what == 'max' ):
         df = df.append(data_by_park[ data_by_park.observations == data_by_park.observations.max() ] ).reset_index(drop=True)
      else:
         print( 'Error, select min or max in what' )
         return None
   return(df.reset_index(drop=True))
```

Using the former function to show the least observed species in each park. The least observed of all species in all the parks is at Bryce national Park, the Corydalis aurea, a vascualr plant has just 9 observations and is has a "No intervention" status. Similar case with the Collomia tinctoria at Great Smoky Mountains with just 10 observations and has a "No intervention" status.
To check the reazon of the "No intervention" status for this species let's check out the number of observations in the others parks.

```python
least_observed = get_min_max_observations(data, 'min')
least_observed.to_csv('E:/CursoDataScience/python/biodiversity-starter/biodiversity_starter/table01.csv', index=False)
least_observed
```
